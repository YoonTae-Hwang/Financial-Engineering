```python
import numpy as np
import scipy.stats as scs
import statsmodels.api as sm
import math
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def get_geobrown_paths(S0 = 100, r = 0.05, vol = 0.2, T = 1.0, M = 50 , N = 2500):
    
        """
        ==== input =====
        S0 : initial price
        T  : time to maturity
        r  : short interest rate (constant)
        vol : volatility (constant)
        M : time interval (constant) 
        N : number of simulation paths (it can make using length pd.DatetimeIndex )

        ==== output =====
        geo brwaon path
        """
        
        
        dt = float(T) / M  # 50 weeks
        paths = np.zeros((M + 1, N), np.float64)
        paths[0] = S0
        
        for t in range(1, M + 1):
            rand = np.random.standard_normal(N)
            rand = (rand - rand.mean()) / rand.std()
            paths[t] = paths[t-1] * np.exp((-r - 0.5 * vol ** 2) *dt + vol * np.sqrt(dt) * rand) 
        return paths
```

```python
paths = get_geobrown_paths()
plt.plot(paths[:, :10])
plt.grid(True)
```
